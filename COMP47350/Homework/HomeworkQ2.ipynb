Potential Problems:
- Age had a significant number of zero values which greatly impacted the descriptive statistics. To have a contract at the age of zero is impossible. Due to this, all zero values were replaced with NaN values such that they are now not considered when the descriptive statistics are being calculated.
- Customer ID had to be changed to a category type. While this dataframe column is populated with integers, it could be populated with integer, string, or integer and string values. It is not a number that will be used for calculations in any way
- Region Type had synonyms for the same value which produced a bar chart that was hard to analyse. I renamed the column attributes so that there were no synonyms such that 't' became 'town', 's' became 'suburban', and 'r' became 'rural'.
- Credit Card On Account also had synonyms. They were changed such that all positive values became 'true' and all negative values became 'false'.
- I think that the Income column should have a scale from 1 to 10. If there were a scenario where it was needed to divide by the income grade, it is impossible to divide by 0. 
- Current Handset Price column had a significant number of zero values that would impact the descriptive statistics. Considering everyone has a phone contract here, they must have a phone. The zero values could represent customers who purchased their phones elsewhere (such as online or with another phone service provider) or else where the company doesn't know the value of the customer's current handset. These zero values are not relevant and such were replaced with NaN values so they would not influence the calculation fo the descriptive statistics.
- Bill Amount Change % column had zero values which were not relevant. If the values stayed the same then it would not be possible to calculate the bill change percentage accurately. A change of zero percent is not a change, but if it's included as a value in the mean, for example, it would impact the validity of the value.
- Number of Retention Calls and No. Retention Offers Accepted both had a large number of zero values. These values would have been statistically insignificant should the zeros not have been changed to NaN. And they reveal a very significant finding that about 50% of retention calls result in customers resigning contracts instead of churning.
