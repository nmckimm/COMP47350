```python
# associating duplicates with their synonyms
df['occupation'].replace('', 'unknown', inplace=True)
df['regionType'].replace('', 'unknown', inplace=True)
df['regionType'].replace('s', 'suburban', inplace=True)
df['regionType'].replace('r', 'rural', inplace=True)
df['regionType'].replace('t', 'town', inplace=True)
df['creditCard'].replace('t', 'true', inplace=True)
df['creditCard'].replace('yes', 'true', inplace=True)
df['creditCard'].replace('f', 'false', inplace=True)
df['creditCard'].replace('n', 'false', inplace=True)
```

```python
## replacing values with NaN so they do not influence average, mean, etc.
df['age'] = df['age'].replace(0 , np.nan) # age 0 does not exist
df['currentHandsetPrice'] = df['currentHandsetPrice'].replace(0 , np.nan) # everyone has a handset in this csv
df['avgBill'] = df['avgBill'].replace(0 , np.nan) # bill cannot be 0 unless the exact amount of the bill is refunded
df['billAmountChangePct'] = df['billAmountChangePct'].replace(0 , np.nan) # bill won't be 0, 104 zero values would skew results
df['numRetentionCalls'] = df['numRetentionCalls'].replace(0 , np.nan) # if there are over 900 zero values, it will be impossible to compare numRetentionCalls to numRetentionOffersAccepted
df['numRetentionOffersAccepted'] = df['numRetentionOffersAccepted'].replace(0, np.nan)

## NaN values have to be float - cannot convert back to int64
```
