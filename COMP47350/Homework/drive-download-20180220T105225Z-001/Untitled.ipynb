```python
df.columns = df.columns.str.strip()
```

```python
df['occupation'].replace('', np.nan, inplace=True)
df.dropna(subset=['occupation'], inplace=True)
```

```python
## Finding cardinality of attributes
print('Unique Occupations: ', pd.unique(df.occupation.str.strip().ravel()), '\n')
print('Unique Region Types: ', pd.unique(df.regionType.str.strip().ravel()), '\n')
print('Unique Marriage Status: ', pd.unique(df.marriageStatus.str.strip().ravel()), '\n')
print('Unique Children: ', pd.unique(df.children.str.strip().ravel()), '\n')
print('Unique Smart Phone: ', pd.unique(df.smartPhone.str.strip().ravel()), '\n')
print('Unique Credit Rating: ', pd.unique(df.creditRating.str.strip().ravel()), '\n')
```

```python
df['regionType'] = df['regionType'].replace({'s': 'suburban', 'r': 'rural', 't': 'town'})
pd.unique(df['regionType'].str.strip().ravel())
```
